#### Change date columns format from object to datetime

```python
# Packages for date conversions for calculating trip durations
from datetime import datetime
from datetime import date
from datetime import timedelta
```

```python
type(df['starttime'][0])
```

```python
# Get the date
df['date'] = df['starttime'].apply(lambda date: date[:10]) #take 0-10 which is Date str
```

```python
# Create and Convert 'Date' column to datetime format
df['date'] = pd.to_datetime(df['date'], format='%Y-%m-%d')
```

```python
# Check names in the columns
df.columns
```

#### Re-arrange column

```python
rename_column = {
    'tripduration':'trip_duration',
    'starttime' : 'start_time',
    'stoptime' : 'stop_time',
    'usertype': 'user_type',
    'bikeid': 'bike_id',
    'gender' : 'sex'}
```

```python
df.rename(columns=rename_column, inplace=True)
```

```python
df.columns.values
```

```python
column_names = ['user_type', 'bike_id', 'trip_duration','date','start_time', 'start_station_id', 
                'start_station_name', 'start_station_latitude','start_station_longitude', 'stop_time', 
                'end_station_id', 'end_station_name', 'end_station_latitude', 'end_station_longitude',
                'birth_year', 'sex', 'customer_plan']

# Reorder the columns using `reindex()`
df = df.reindex(columns=column_names)

# Display the DataFrame with the updated column order
df.head()
```

#### Change the starttime and stoptime data from object into datetime

```python
type(df['start_time'][0])
```

```python
# Convert `start` to datetime format
df['start_time']=pd.to_datetime(df['start_time'])

# Convert `stop` to datetime format
df['stop_time']=pd.to_datetime(df['stop_time'])

# Compute `trip_duration` in minutes
df["trip_duration"] = df["trip_duration"]/60


# Get the year range
min_year = df['start_time'].dt.year.min()
max_year = df['stop_time'].dt.year.max()

print(f"The year range in the DataFrame is from {min_year} to {max_year}.")
```

```python
df.info()
```

#### Acquire data between 2014 and November 2015

```python
# Filter the DataFrame to include only the data between 2014 and 2015
#start_date = pd.to_datetime('2014-01-01')
#end_date = pd.to_datetime('2015-11-30')

# Filter the DataFrame to include only the data between 2014 and 2015
df = df[(df['start_time'] >= '2014-01-01') & (df['stop_time'] <= '2015-11-30')]
```

```python
# Get the year range
min_year = df['start_time'].dt.year.min()
max_year = df['start_time'].dt.year.max()

print(f"The year range in the DataFrame is from {min_year} to {max_year}.")
```

```python
# Save the filtered DataFrame to a CSV file
df.to_csv('NYC_trip_data_2014_2015.csv', index=False)
```
